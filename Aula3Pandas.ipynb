```python
#criando o dataframe
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#quantidade de linhas e colunas
df.shape
```

```python
#verificando dados
df.dtypes
```

```python
#qual a Receita total?
df['Valor Venda'].sum()
```

```python
#qual o custo total?
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])#criando a coluna de custo
```

```python
df.head(1)
```

```python
#qual o custo total?
round(df['Custo'].sum(), 2)
```

```python
#agora temos a receita e custo e o total, podemos achar o lucro total
#Vamos criar uma coluna de Lucro que será Receita - Custo
df['Lucro'] = df['Valor Venda'] - df['Custo']
```

```python
df.head(1)
```

```python
#total lucro
round(df['Lucro'].sum(), 2)
```

```python
#criando uma coluna com o total de dias para enviar o produto
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda'] 
```

```python
df.head(1)
```

## **Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_envio em númerica**

```python
#extraindo apenas os dias
df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
#verificando o tipo de coluna d eenvio
df['Tempo_envio'].dtypes
```

```python
#media do tempo de envio por marca
df.groupby('Marca')['Tempo_envio'].mean()
```
